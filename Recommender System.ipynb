```python
Ratings=pd.read_csv('ratings.csv')
Ratings.describe()
```

```python
df2=pd.merge(df,Ratings,on='movieId')
```

```python
for i in df2.index:
    record= df2.iloc[i,:]
    record= record.dropna()
df2
```

```python
req_data=[]
a=input("Enter any genre: ")
for i in df2.index:
    movie=df2.loc[i,'title']
    row=df2.loc[i,'genres']
    rate=df2.loc[i,'rating']
    if type(row) is not float:
        if a in row:
            req_data.append({'title':movie,'genres':row,'rating':rate})
req_data=pd.DataFrame(req_data)
final_req_data=pd.DataFrame(req_data.groupby(['title','genres'])['rating'].mean())
#req_data
final_req_data=final_req_data.dropna()
print(type(rate))
final_req_data.sort_values(by='rating',ascending=False).head(50)
```

```python
movie_matrix = df2.pivot_table(index='userId', columns='title', values='rating')
#movie_matrix.dropna()
```

```python
moviedf=pd.DataFrame(movie_matrix)
```

```python
film=input("Enter a movie")
til_rating2 = moviedf[film]
similar=moviedf.corrwith(til_rating2)
corr = pd.DataFrame(similar, columns=['Correlation'])
corr.dropna(inplace=True)
corr.head(50)
```
